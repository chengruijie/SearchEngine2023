```python
import os
from bs4 import BeautifulSoup
import xml.etree.ElementTree as ET
```

```python
case_id = []
case_type = []
```

```python
cnt = 0
for root_dir, dirs, files in os.walk('../resources/static'):
    for file in files:
        cnt += 1

idx = 0
for root_dir, dirs, files in os.walk('../resources/static'):
    for file in files:
        # 进度条
        idx += 1
        print(f'\r{idx}/{cnt}, {file}', end='')
        if file.endswith('.xml'):
            try:
                tree = ET.parse(os.path.join(root_dir, file))
                root = tree.getroot().find('QW').find('WS')
                if root is None:
                    continue
            except:
                continue
            # 案号
            AH = root.find('AH')
            if AH is not None:
                case_id.append(AH.get('value'))
            # 案件类别
            AJLB = root.find('AJLB')
            if AJLB is not None:
                case_type.append(AJLB.get('value'))
```
